### DBScan Clustring
##### Work on Non-Linearly Separable dataset
##### Use for Outlier Detection

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y =make_moons(n_samples=250 ,noise=0.05)
```

```python
df ={'feature 1':x[:,0],'feature 2':x[:,1]}
```

```python
dataset =pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x ='feature 1', y='feature 2',data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db =DBSCAN(eps=0.2,min_samples=5)
dataset['Predict']=db.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x ='feature 1', y='feature 2',data=dataset,hue='Predict')
plt.show()
```
