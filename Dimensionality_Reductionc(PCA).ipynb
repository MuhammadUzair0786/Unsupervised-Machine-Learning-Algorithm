##  Dimensionality reduction (Pricipal Component Analysis)

```python
import pandas as pd
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
```

```python
dataset =load_digits()
dataset.keys()
```

```python
dataset.data.shape
```

```python
dataset.data[0].reshape(8,8)
```

```python
plt.gray()
plt.matshow(dataset.data[2].reshape(8,8))
```

```python
dataset.target[2]
```

```python
df =pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
```

```python
df.head(3)
```

```python
df.describe()
```

```python
x =df
y =dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled =scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train,y_test =train_test_split(x_scaled,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
x.head()
```

```python
from sklearn.decomposition import PCA
pca =PCA(0.95)
x_pca =pca.fit_transform(x)
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
x_train_pca ,x_test_pca ,y_train,y_test =train_test_split(x_pca,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train)
```

```python
model.score(x_test_pca,y_test)*100
```

### Component Method of PCA


```python
from sklearn.decomposition import PCA
pca =PCA(n_components=10)
x_pca =pca.fit_transform(x)
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
x_train_pca ,x_test_pca ,y_train,y_test =train_test_split(x_pca,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train)
```

```python
model.score(x_test_pca,y_test)*100
```
